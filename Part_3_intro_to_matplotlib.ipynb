### 3.12 Select Columns with Methods

Although column selection is usually done directly with the indexing operator, there are some DataFrame methods that facilitate their selection in an alternative manner. select_dtypes and filter are two useful methods to do this.

Let's see the number of columns of each data type available

````
canada_df.dtypes.value_counts()
````

**3.12.1 select_dtypes() method:**

Use the select_dtypes method to select only the integer columns:

````
canada_df.select_dtypes(include='int').head()
# This code selects only the columns containing integers and displays the first five rows
````

If we had columns containing both floats and ints, we could select all numeric columns by passing the number parameter...
```
canada_df.select_dtypes(include='number')
```

selecting all columns of strings or mixed data types called objects from the data frame and displaying the first 5 rows

````
canada_df.select_dtypes(include='object').head()
````

**3.12.2 Using The filter() method to select columns:**

An alternative method to select columns is with the filter method. This method is flexible and searches column names (or index labels) based on which parameter is used, such as `like`, `regex` and `item`.

````
# Using like parameter to search for all column names that contain the exact string, 'Co':
canada_df.filter(like='Co').head()
````

The filter method allows columns to be searched through regular expressions with the regex parameter. Here, we search for all columns that have a digit somewhere in their name:

````
# Using regex param to filter all columns with a zero in their name
canada_df.filter(regex='0').head()
````

The filter method comes with another parameter, items, which takes a list of exact column names. This is nearly an exact duplication of the indexing operator, except that a KeyError will not be raised if one or all of the strings does not match any column name

````
canada_df.filter(items=['1980', '1990', '2099']).head()
````

### 3.22 set_index() and reset_index() methods:

Notice that the default index of the dataset is a numeric range from 0 to 194. This makes it very difficult to do a query by a specific country. For example to search for data on Japan, we need to know the corressponding index value.

This can be fixed very easily by setting the 'Country' column as the index using `set_index()` method.

The opposite of set is reset. So to reset the index, we can use `reset_index()` method.

After setting the index, to remove the name from the index,  we can do:-<br>
`canada_df.index.name = None`

Let's make a copy of `canada_df` so that the original is not altered

````
canada_copy = canada_df.copy(deep=True)
````

````
canada_copy.set_index('Country', inplace=True, drop=True)
canada_copy.head()
````

Let's remove the label on the index

````
canada_copy.index.name = None
canada_copy.head()
````

**3.23 Example:** 

Let's view the number of immigrants from Japan (row 87) for the following scenarios:

The full row data (all columns)<br>
For year 2013<br>
For years 1980 to 1985

````
# To get the full row of Japan with all columns

canada_copy.loc[['Japan']]  # we use a list to make the result a data frame
````

````
# Let's repeat with the iloc[] command passing the numeric index

canada_copy.iloc[[87]]  # we use a list to make the result a data frame
````

Let's now select the result for Japan for only year 2013 (column at position 36)

````
canada_copy.loc['Japan', '2013']
````

````
# Let's do same with iloc[]

canada_copy.iloc[87, 36]
````

Let's now select the result for Japan for years 1980 to 1985

````
canada_copy.loc['Japan', '1980':'1985'].to_frame()
# slicing it and using the to_frame() method
````

**3.24 Slicing Rows lazily**

The .iloc and .loc indexers were used to select subsets of both Series and DataFrames in either dimension. A shortcut to select the rows exists with just the indexing operator itself. <br>This is just a shortcut to show additional features of pandas, but the primary function of the indexing operator is actually to select DataFrame columns. If you want to select rows, it is best to use .iloc or .loc, as they are unambiguous.
